# read all the images and convert them into array

```python
training_Data = [] ## data array

def create_training_Data():
    for category in Classes:
        path = os.path.join(Datadirectory, category)
        class_num = Classes.index(category) ## 0 1, ## Label
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path, img))
                new_array = cv2.resize(img_array, (img_size, img_size))
                training_Data.append([new_array, class_num])
            except Exception as e:
                pass

create_training_Data()
```

```python
print(len(training_Data))
```

```python
#we are just using training and deploying for real time webcame demo
```

```python
import random 
random.shuffle(training_Data)
```

```python
X = [] ## Data/feature
y = [] ## Label
```

```python

for features,label in training_Data:
    X.append(features)
    y.append(label)
```

```python
X = np.array(X).reshape(-1, img_size, img_size, 3) ## converting it to 4 dimension, (-1 means end)
```

```python
X.shape ## (no. of images, other 3 is size the image)
```

```python
#normalize the data
X = X/255.0

## we can also use scikit learn (but we are basic that's why divide by 255.0)
```

# Deep Learning model for Training - Transfer Learning

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = tf.keras.applications.MobileNetV2() ## pre-trained model
```

```python
model.summary() ## we changed everything to 224 initilly becz mobileNet accept only Model: "mobilenetv2_1.00_224"
```

# Transfer Learning - Tuning, weights will start from last check point

```python
base_input = model.layers[0].input ## my input
```

```python
base_output = model.layers[-2].output
```

```python
base_output
```

```python
final_output = layers.Dense(128)(base_output) ## adding new layer, after the output of global pooling layer
final_ouput = layers.Activation('relu')(final_output) ## activation function
final_output = layers.Dense(64)(final_ouput) 
final_ouput = layers.Activation('relu')(final_output)
final_output = layers.Dense(7, activation='softmax')(final_ouput) ## why 7?...becz my classes are 7 (classification layer)
```

```python
final_output
```

```python
new_model = keras.Model(inputs = base_input, outputs = final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(loss="sparse_categorical_crossentropy", optimizer = "adam", metrics = ["accuracy"])
```

```python
new_model.fit(X,y, epochs=25)
## if are not getting good accuracy then increase the epochs
```

```python
new_model.save('emotion_model.h5')
## we are doing transfer model where running already exist model to our time.
## Now we going to save our model to use in future => Transfer Learing
```
